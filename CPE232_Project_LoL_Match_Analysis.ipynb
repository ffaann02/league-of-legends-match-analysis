#1. Data Explanation
ชุดข้อมูล "League of Legends" เป็นการเก็บสถิติของการแข่งขันเกม LoL ของนักแข่ง eSport ในลีกระดับมืออาชีพ โดยระบุข้อมูลที่สำคัญที่เกิดขึ้นในเกม เช่น จำนวนป้อม, จำนวนการโค่นฝ่ายตรงข้าม, ความต่างของเงินสะสมในทีม เป็นต้น

```python
import pandas as pd
pd.set_option('display.max_columns', None)

df = pd.read_csv("/content/LeagueofLegends.csv")

df.head(2)
```

ในโปรเจคครั้งนี้ กลุ่มของนักศึกษาได้มีความสนใจในการนำข้อมูลสถิติการเล่นที่เกิดขึ้นในระหว่างเกม
มาสร้างเป็นโมเดลเพื่อทำนายว่าผู้เล่นในทีมจะชนะหรือแพ้ (Classification) และจัดกลุ่มสไตล์การเล่นของแต่ละทีม (Clustering) โดยลักษณะข้อมูลที่ใช้มีดังนี้
- แต่ละ row แทนการแข่งขัน 1 Match โดยระบุข้อมูลลีก, ปี, ฤดูกาล
- เก็บข้อมูลทีมฝ่ายน้ำเงินและฝ่ายแดงใน row เดียวกันโดยใช้ตัวอักษรนำหน้าแทน 'r' คือ red (ฝ่ายแดง) และ 'b' คือ blue (ฝ่ายน้ำเงิน)
- เก็บข้อมูลการเงินสะสมภายในทีมในรูปแบบ Array โดยแต่ละ index แทนด้วยระยะเวลาที่ผ่านไปในเกมใน 1 นาที / index
- เก็บข้อมูลการโค่นศัตรู, ทำลายป้อมปราการ, ทำลายฐานทัพ, สังหารมังกร, สังหารสัตว์ป่าในตำนาน ในรูปแบบของ Array 2 มิติ โดยแต่ละ index แทนการกระทำที่เกิดขึ้น 1 ครั้ง ซึ่งภายในเก็บข้อมูลรายละเอียดการกระทำ เช่น คนสังหาร, เลนที่โดนทำลายป้อม และข้อมูลอื่น ๆ

#2. Data Preparation Process and Results
ด้วยลักษณะการเก็บข้อมูลแบบรวมทั้ง 2 ทีม (แดงและน้ำเงิน) ภายใน row ด้วยกัน และการเก็บข้อมูลรายละเอียดบางอย่างใน Array ทำให้ใช้งานได้ไม่สะดวก และค่อนข้างยุ่งยากในการวิเคราะห์ โดยสามารถแก้ไขและจัดรูปแบบข้อมูลใหม่ให้นำไปใช้ต่อได้ง่ายด้วยขั้นตอนดังต่อไปนี้

##2.1 แยกข้อมูลให้อยู่ในรูป 1st Normalized Form (1NF) หรือไม่มี Repeating Groups ใน row เดียวกัน
โดยปัญหาในปัจจุบันคือ ใน row เดียวกันเก็บข้อมูลทั้งทีมฝั่งสีแดงและน้ำเงิน เราจึงควรแยกออกมาเก็บข้อมูลเป็นสถิติทีมในแต่ละ row แทน

```python
df_separated = df.copy()
df_separated.head(2)
```

แบ่ง Column เป็น 2 แบบ: แบบใช้ร่วมกัน (common) และ แบบแยกทีม (team)

```python
common_columns = ['League', 'Year', 'Season', 'Type','gamelength', 'golddiff', 'Address']

col_blue = common_columns + ['blueTeamTag', 'bResult','goldblue', 'bKills', 'bTowers', 'bInhibs',
'bDragons','bBarons', 'bHeralds', 'blueTop', 'blueTopChamp','goldblueTop', 'blueJungle', 'blueJungleChamp',
'goldblueJungle','blueMiddle', 'blueMiddleChamp', 'goldblueMiddle','blueADC', 'blueADCChamp',
'goldblueADC','blueSupport', 'blueSupportChamp', 'goldblueSupport','blueBans']


col_red = common_columns + ['redTeamTag', 'rResult', 'goldred', 'rKills', 'rTowers', 'rInhibs',
'rDragons', 'rBarons', 'rHeralds', 'redTop', 'redTopChamp','goldredTop', 'redJungle', 'redJungleChamp',
'goldredJungle','redMiddle', 'redMiddleChamp', 'goldredMiddle','redADC',
'redADCChamp', 'goldredADC','redSupport', 'redSupportChamp','goldredSupport' , 'redBans']
```

```python
df_blue = df_separated[col_blue]
df_red = df_separated[col_red]
# ตรวจสอบว่าแบ่งออกมาจาก Parent Data Frame แล้วจำนวน Column ยังคงเท่าเดิมหรือไม่ (ตรวจสอบว่าแบ่งถูกต้องหรือไม่)
print(f'Parent Col:{len(df_separated.columns)} = share:{len(common_columns)} + blue:{len(df_blue.columns) - len(common_columns)} + red:{len(df_red.columns) - len(common_columns)}')

display(df_blue.head(1))
display(df_red.head(1))
```

```python
# ใส่ Index ในรูป 'Bi' และ 'Ri' สำหรับทีมน้ำเงินและแดง โดย i คือลำดับ Match ที่เล่น เช่น 'B1', 'R1' คือคู่แข่งในแต่ละฝั่งของ Match ที่ 1
df_blue = df_blue.assign(Index=lambda x: 'B' + (x.index + 1).astype(str))
df_red = df_red.assign(Index=lambda x: 'R' + (x.index + 1).astype(str))

# เปลี่ยนชื่อ Index เป็น matchID
df_blue.rename(columns={'Index': 'matchID'}, inplace=True)
df_red.rename(columns={'Index': 'matchID'}, inplace=True)

# ขยับ matchID ไปเป็น column แรก
df_blue = df_blue[['matchID'] + [col for col in df_blue if col != 'matchID']]
df_red = df_red[['matchID'] + [col for col in df_red if col != 'matchID']]
```

```python
#ทดลองแสดงผล Data Frame เพื่อดูว่าถูกต้องตามที่ต้องการหรือไม่
display(df_blue.set_index('matchID').head(1))
display(df_red.set_index('matchID').head(1))
```

##2.2 Drop ข้อมูลที่ไม่มีนัยสำคัญในการวิเคราะห์
เนื่องจากเกม LoL เป็นการแข่งขันประเภททีม 5 ต่อ 5 การนำข้อมูลประเภทที่ใช้ระบุผู้เล่นรายบุคคลและสถิติส่วนตัว นั้นอาจไม่มีผลกระทบมากต่อการวิเคราะห์มาก นักศึกษาจึงเลือกใช้ข้อมูลสถิติของทีมเป็นหลัก

ข้อมูลใน Column index ที่ 17 ถึงสุดท้าย เป็นข้อมูลส่วนตัวของผู้เล่นในทีม เช่น ตำแหน่ง, ฮีโร่ที่ใช้, การเงินส่วนตัว, การแบนตัวละครก่อนเริ่มเกม ซึ่งไม่ได้อยู่ในขอบเขตของโปรเจคในครั้งนี้ ที่มุ่งเน้นวิเคราะห์จากสถิติรวมของทีม จึงตัดสินใจ Drop เพื่อลดจำนวน Parameter ที่ใช้ในขั้นตอนถัด ๆ ไป

```python
columns_to_drop_blue = df_blue.columns[17:]
columns_to_drop_red = df_red.columns[17:]

df_blue.drop(columns=columns_to_drop_blue, inplace=True)
df_red.drop(columns=columns_to_drop_red, inplace=True)
```

Drop ข้อมูล Address ซึ่งเป็น Url Link อ้างอิงสถิติ ซึ่งไม่จำเป็นในการใช้วิเคราะห์ข้อมูลในโปรเจคนี้

```python
df_blue.drop(columns=['Address'],inplace=True)
df_red.drop(columns=['Address'],inplace=True)
```

```python
#ทดลองแสดงผล Data Frame เพื่อดูว่าถูกต้องตามที่ต้องการหรือไม่
display(df_blue.set_index('matchID').head(1))
display(df_red.set_index('matchID').head(1))
```

##2.3 แปลงข้อมูลสถิติทั้ง 2 ทีมเป็น Data Frame เดียวกัน

เปลี่ยนชื่อ Column ที่ระบุทีม เช่น ขึ้นต้นด้วย 'r' หรือ 'red' ให้อยู่ในรูปที่สามารถใช้ชื่อเดียวกัน จะได้รวม Data Frame กันได้

```python
column_mapping = {
    0: 'TeamName',
    1: 'Result',
    2: 'Gold',
    3: 'Kills',
    4: 'Towers',
    5: 'Inhibitors',
    6: 'Dragons',
    7: 'Barons',
    8: 'Heralds',
}
```

```python
df_blue.columns = df_blue.columns[:7].tolist() + [column_mapping.get(i, col) for i, col in enumerate(df_blue.columns[7:])]
df_red.columns = df_red.columns[:7].tolist() + [column_mapping.get(i, col) for i, col in enumerate(df_red.columns[7:])]
```

```python
df_blue.rename(columns={'gamelength': 'GameLength'}, inplace=True)
df_red.rename(columns={'gamelength': 'GameLength'}, inplace=True)
```

```python
df_blue['Result'] = df_blue['Result'].replace({1: 'Win', 0: 'Lose'})
df_red['Result'] = df_red['Result'].replace({1: 'Win', 0: 'Lose'})
```

```python
#ทดลองแสดงผล Data Frame เพื่อดูว่าถูกต้องตามที่ต้องการหรือไม่
display(df_blue.set_index('matchID').head(1))
display(df_red.set_index('matchID').head(1))
```

เรียงข้อมูลโดยสลับเป็นคู่ ๆ ในแต่ละแถว เพื่อให้เปรียบเทียบได้ง่ายในระหว่างกระบวนการทำขั้นตอนต่าง ๆ
ตัวอย่าง
- B1: ทีมฝั่ง Blue ในเกมที่ 1
- R1: ทีมฝั่ง Red ในเกมที่ 1

```python
index_blue = pd.MultiIndex.from_product([['B'], range(1, len(df_blue)+1)], names=['prefix', 'matchID'])
index_red = pd.MultiIndex.from_product([['R'], range(1, len(df_red)+1)], names=['prefix', 'matchID'])

df_blue.index = index_blue
df_red.index = index_red

df_merge = pd.concat([df_blue, df_red])
df_merge = df_merge.sort_index(level='matchID')
```

```python
display(df_merge.set_index('matchID').head(2))
```

ตรวจสอบว่าการรวมข้อมูลของการเล่นทั้ง 2 ฝ่ายตรงข้ามกันใน Data Frame เดียวกันนั้นถูกต้องหรือไม่ ซึ่งข้อมูลควรจะมีการกระจายตัวแบบ 50:50 ในฝั่งทีม Blue:Red และ 50:50 ในผลการแข่ง Win:Lose

```python
df_B = df_merge[df_merge['matchID'].str.startswith('B')]
df_R = df_merge[df_merge['matchID'].str.startswith('R')]

print('จำนวน rows ที่ matchID เป็น B และ R')
print(f'B:{len(df_R)} R:{len(df_R)}')

df_W = df_merge[df_merge['Result']=='Win']
df_L = df_merge[df_merge['Result']=='Lose']

print('\nจำนวน rows')
print(f'Win:{len(df_W)} Lose:{len(df_L)}')
```

บันทึก Data Frame ที่รวมข้อมูลแล้วในรูปแบบไฟล์ CSV เพื่อให้ง่ายในการทำงานต่อร่วมกันภายในกลุ่ม

พบปัญหาคือ ข้อมูลที่คาดว่าเป็น Array แท้จริงแล้วเป็น String ซึ่งทำให้การเข้าถึงแต่ละ Index นั้นไม่ถูกต้อง เช่น ในกรณีที่ต้องการ index 0 ควรจะได้ข้อมูลในตำแหน่ง 0 แต่กลับได้ '[' เราจึงควรแปลงให้อยู่ในรูปแบบ Array เสียก่อน

```python
selected_columns_indices = [5] + list(range(8, len(df_new.columns)))

for col_index in selected_columns_indices:
    first_element = df_new.iloc[0, col_index][0]
    print(f"{first_element}")
```

```python
import ast

df_new_array = df_new.copy()
for col_index in selected_columns_indices:
    df_new_array.iloc[:, col_index] = df_new_array.iloc[:, col_index].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
df_new_array_1 = df_new_array.copy()
```

สังเกตได้ว่าข้อมูลใน Index 0 ของ Column ที่เป็น Array ได้แสดงข้อมูลที่เก็บภายในอย่างถูกต้องแล้ว

```python
for col_index in selected_columns_indices:
    first_element = df_new_array_1.iloc[0, col_index]
    print(f"{first_element}")
```

```python
df_new_array_1.head(2)
```

'golddiff' เป็นข้อมูลความต่างของการเงินสะสมรวมของทีมกับทีมศัตรูตรงข้าม โดยรายละเอียดดังนี้
- ในแต่ละ Index ที่เพิ่มแทนระยะเวลา 1 นาที
- 0 คือไม่มีความต่างของเงิน
- เลขมีค่าเป็น + คือการเงินนำทีมตรงข้าม
- เลขมีค่าเป็น - คือการเงินตามหลังทีมตรงข้าม

ปัญหาในปัจจุบันคือ 'golddiff' นั้นคำนวณจาก Gold(Blue) - Gold(Red) ซึ่งใช้วิธีการนี้ทั้ง 2 ฝั่ง แต่ในความเป็นจริงแล้ว ทีมฝั่ง Red ควรจะมี 'golddiff' คำนวณจาก Gold(Red) - Gold(Blue) ดังนั้นเราจึงใช้วิธีนำค่า golddiff ในแถวของทีม Red มา Inverse ได้ดังนี้

```python
print(df_new_array_1.loc['B1','golddiff'])
print(df_new_array_1.loc['R1','golddiff'])
```

คำนวณค่า 'golddiff' ในฝั่งทีมสีแดงได้อย่างง่ายด้วยการนำค่า 'golddiff' ในฝั่งทีมน้ำเงินมา Inverse

```python
red_indices = df_new_array_1.index[df_new_array_1.index.str.startswith('R')]
df_new_array_1.loc[red_indices, 'golddiff'] = df_new_array_1.loc[red_indices, 'golddiff'].apply(lambda x: [-val for val in x])

print(df_new_array_1.loc['R1','golddiff'])
```

```python
df_convert_Gdiff = df_new_array_1.copy()
```

```python
result_index = 7

df_convert_Gdiff.insert(result_index,'golddiff',df_convert_Gdiff.pop('golddiff'))
df_convert_Gdiff.head(2)
```

แปลงข้อมูล 'golddiff' เป็น Column: 'minGoldDiff' และ 'maxGoldDiff'

```python
golddiff_index = df_convert_Gdiff.columns.get_loc('golddiff')
def extract_min_max(values):
    min_value = min(values)
    max_value = max(values)
    return min_value, max_value
```

```python
df_convert_Gdiff[['minGoldDiff', 'maxGoldDiff']] = df_convert_Gdiff['golddiff'].apply(lambda x: pd.Series(extract_min_max(x)))
```

```python
df_convert_Gdiff.drop(columns=['golddiff'], inplace=True)

df_convert_Gdiff.insert(golddiff_index + 1, 'minGoldDiff', df_convert_Gdiff.pop('minGoldDiff'))
df_convert_Gdiff.insert(golddiff_index + 2, 'maxGoldDiff', df_convert_Gdiff.pop('maxGoldDiff'))

df_convert_Gdiff.head(2)
```

แปลงข้อมูล 'Gold' เป็น 'MidGameGold' และ 'TotalGold'

```python
gold_index = df_convert_Gdiff.columns.get_loc('Gold')
```

```python
df_convert_Gdiff['MidGameGold'] = df_convert_Gdiff['Gold'].apply(lambda x: x[len(x) // 2])
df_convert_Gdiff['TotalGold'] = df_convert_Gdiff['Gold'].apply(lambda x: x[-1])
```

```python
df_convert_Gdiff.drop(columns=['Gold'], inplace=True)

df_convert_Gdiff.insert(gold_index + 1, 'MidGameGold', df_convert_Gdiff.pop('MidGameGold'))
df_convert_Gdiff.insert(gold_index + 2, 'TotalGold', df_convert_Gdiff.pop('TotalGold'))

df_convert_Gdiff.head(2)
```

```python
min_gold_diff_index = df_convert_Gdiff.columns.get_loc('minGoldDiff')
max_gold_diff_index = df_convert_Gdiff.columns.get_loc('maxGoldDiff')

if max_gold_diff_index > min_gold_diff_index + 1:
    df_convert_Gdiff.insert(min_gold_diff_index + 1, 'maxGoldDiff', df_convert_Gdiff.pop('maxGoldDiff'))
```

เพิ่ม Column 'FirstKill' โดยเก็บข้อมูลเป็น Boolean เพื่อมีประโยชน์ในขั้นตอนถัดไป เช่น การดูแนวโน้มว่าทีมที่สังหารได้ก่อน จะมีผลต่อการชนะแพ้หรือไม่

```python
def check_first_kill(row, df):
    match_id = row.name
    team_a = None
    team_b = None
    if match_id.startswith('B'):
        team_a = df.loc[match_id, 'Kills']
        team_b = df.loc['R' + match_id[1:], 'Kills']
    elif match_id.startswith('R'):
        team_a = df.loc[match_id, 'Kills']
        team_b = df.loc['B' + match_id[1:], 'Kills']

    if team_a and not team_b:
        return True
    elif team_b and not team_a:
        return False
    elif team_a and team_b:
        return team_a[0][0] <= team_b[0][0]
    else:
        return None
```

```python
df_win = df_eda[df_eda['Result'] == 'Win']
df_lose = df_eda[df_eda['Result'] == 'Lose']

numeric_columns = df_eda.select_dtypes(include='number').columns
fig, axes = plt.subplots(nrows=len(numeric_columns), ncols=1, figsize=(12, 16))

for i, column in enumerate(numeric_columns):
    axes[i].hist(df_win[column], bins=20, alpha=0.5, label='Win')
    axes[i].hist(df_lose[column], bins=20, alpha=0.5, label='Lose')
    axes[i].set_title(column)
    axes[i].legend()

plt.tight_layout(pad=2.0)
plt.show()
```

จากการพล็อตกราฟสังเกตได้ว่าทีมที่ชนะ จะมีลักษณะการกระจายตัวของกลุ่มข้อมูลที่ Mean สูงกว่าทีมที่แพ้ เพราะประสิทธิภาพในการเล่นที่ดีกว่า เช่น minGoldDiff ที่กระจุกตัวอยู่บริเวณ 0 แสดงว่าทีมที่ชนะส่วนใหญ่จะมีการเงินนำทีมศัตรูเสมอ แต่ในกรณีของข้อมูลการเงินระหว่างเกม (MidGameGold) ที่มีการกระจายตัวในลักษณะค่อนข้างคล้ายกัน และอยู่ในช่วงใกล้กัน แสดงให้เห็นว่าการเงินระหว่างเกมไม่สามารถตัดสินผู้ชนะหรือแพ้ได้ เป็นต้น

```python
df_win = df_eda[df_eda['Result'] == 'Win']
df_lose = df_eda[df_eda['Result'] == 'Lose']

stat_columns = ['GameLength', 'TotalKills', 'TotalAssists', 'TotalDeath']
gold_columns = ['minGoldDiff', 'maxGoldDiff', 'MidGameGold', 'TotalGold']
objective_columns = ['Towers', 'Inhibitors', 'Dragons', 'Barons', 'Heralds']

desc_win = df_win.describe().add_prefix('Win_')
desc_lose = df_lose.describe().add_prefix('Lose_')

desc_combined = pd.concat([desc_win, desc_lose], axis=1)

def reorganize_stats(stat_columns, desc_combined):
    columns = []
    for col in stat_columns:
        columns.append(f'Win_{col}')
        columns.append(f'Lose_{col}')
    return desc_combined[columns]

stat_group = reorganize_stats(stat_columns, desc_combined)
gold_group = reorganize_stats(gold_columns, desc_combined)
objective_group = reorganize_stats(objective_columns, desc_combined)

display("Stat Group:")
display(stat_group)

display("\nGold Group:")
display(gold_group)

display("\nObjective Group:")
display(objective_group)
```

เปรียบเทียบค่าสถิติต่าง ๆ ของทีมชนะและทีมแพ้ เพื่อทดสอบสมมติฐานว่าทีมที่ชนะจะต้องเล่นได้ประสิทธิภาพดีกว่าทีมแพ้ในทุกแง่มุม แล้วจึงตรวจสอบความสัมพันธ์ เช่น Mean ของการสังหาร การช่วยเหลือการสังหารของทีมชนะจะมากกว่าทีมแพ้ และ Mean การเสียชีวิตของทีมชนะจะน้อยกว่าทีมแพ้ เป็นต้น

#4. Modeling
ข้อมูล League of Legends Data ที่นักศึกษาได้ใช้ในโปรเจคครั้งนี้นั้นมีการทำ Model หลายประเภทที่น่าสนใจ ดังนี้
1. Classification: สร้างโมเดลทำนายว่าทีมใดจะชนะหรือแพ้ (Result) โดยใช้โมเดล เช่น Logistic Regression, Decision Trees เป็นต้น
2. Regression: สร้างโมเดลทำนายผลลัพธ์ของตัวเลขค่าเงินในทีม (TotalGold)ซึ่งเป็นค่าตัวเลขต่อเนื่อง (Continious Outcome) ว่าจะประมาณเท่าไหร่เมื่อระยะเวลาของเกมผ่านไปกี่นาที (GameLength) โดยใช้โมเดล เช่น Linear Regression
3. Clustering: สร้างโมเดลจัดกลุ่มของ Match การแข่งขันว่ามีลักษณะอย่างไร เพื่อสำรวจและตรวจสอบ Pattern ของการแข่งขันในภาพรวม ซึ่งไม่ต้องใช้ Defined Lables โดยใช้โมเดล เช่น K-Means Clustering

##4.1 Classification ทำนายว่าทีมจะมีผลชนะหรือแพ้
โดยใช้โมเดลหลายตัวเพื่อเปรียบเทียบประสิทธิภาพ ดังนี้
1. RandomForestClassifier


###4.1.1 แบ่งชุดข้อมูลเป็น Train และ Test
โดยใช้สัดส่วน train:test = 70:30 อย่างเท่า ๆ กัน เพื่อเป็นตัวแปรควบคุมในการทดสอบประสิทธิภาพของโมเดล Classification ต่าง ๆ และแปลงข้อมูลประเภท Categorical ให้เป็น Dummy Variables เช่น การสังหารมังกรตัวแรก (FirstKillDragon) เป็นต้น

```python
df_classify = df_eda.copy()
df_classify.head()
```

```python
from sklearn.model_selection import train_test_split

df_classify.drop(columns=['League', 'TeamName', 'Side', 'Season', 'Type'], inplace=True)
df_classify = pd.get_dummies(df_classify, drop_first=True)

X = df_classify.drop(columns=['Result_Win'])
y = df_classify['Result_Win']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

###4.1.2 ทำนายด้วยโมเดล Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Logistic Regression
log_reg = LogisticRegression(max_iter=10000)
log_reg.fit(X_train, y_train)
y_pred_log_reg = log_reg.predict(X_test)

print("Logistic Regression Classification Report:\n")
print(classification_report(y_test, y_pred_log_reg))
print("-" * 60)
```

###4.1.3 ทำนายด้วยโมเดล Random *Forest*

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("Random Forest Classification Report:\n")
print(classification_report(y_test, y_pred_rf))
print("-" * 60)
```

###4.1.4 ทำนายด้วยโมเดล Support Vector Machine (SVM)

```python
from sklearn.svm import SVC

# Support Vector Machine
svm = SVC()
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)

print("Support Vector Machine Classification Report:\n")
print(classification_report(y_test, y_pred_svm))
print("-" * 60)
```

###4.1.5 ทำนายด้วยโมเดล K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier

# K-Nearest Neighbors
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

print("K-Nearest Neighbors Classification Report:\n")
print(classification_report(y_test, y_pred_knn))
print("-" * 60)
```

###4.1.6 ทำนายด้วยโมเดล Decision Tree

โดยใช้ K-means Clustering Model ในการจัดกลุ่ม Pattern ของลักษณะการเล่นในแต่ละ Match

```python
df_cluster = df_eda.copy()
```

###4.3.1 หาจำนวนของ K กลุ่มที่จะแบ่ง Cluster โดยใช้วิธี Elbow Method

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Drop non-numeric columns from the DataFrame
df_cluster_numeric = df_cluster.select_dtypes(include=[np.number])

# Initialize an empty list to store the within-cluster sum of squares (WCSS)
wcss = []

# Define the range of K values to try
k_values = range(1, 11)

# Calculate WCSS for each K value
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_cluster_numeric)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(8, 6))
plt.plot(k_values, wcss, marker='o', linestyle='-', color='b')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

เลือกทำการ Cluster โดย K = 3 กลุ่ม เพราะเป็นจุดที่มีการหักศอกมากที่สุด

###4.3.2 ใช้โมเดล K-Means ในการทำ Clustering
โดยทำการ Normalize ข้อมูลในแต่ละ Attribute ให้อยู่ใน Scale เดียวกัน

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the data
df_cluster_numeric_normalized = scaler.fit_transform(df_cluster_numeric)

# Fit K-means clustering with normalized data
kmeans_normalized = KMeans(n_clusters=3, random_state=42)
kmeans_normalized.fit(df_cluster_numeric_normalized)

# Get cluster labels for each data point
cluster_labels_normalized = kmeans_normalized.labels_

# Add cluster labels to the original DataFrame
df_cluster['Cluster_Normalized'] = cluster_labels_normalized

# Display the counts of data points in each cluster
print("Counts of data points in each cluster (Normalized):")
print(df_cluster['Cluster_Normalized'].value_counts())

# You can also explore the cluster centers
cluster_centers_normalized = kmeans_normalized.cluster_centers_
print("\nCluster centers (Normalized):")
print(cluster_centers_normalized)
```
